```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation
```

```python
stopwords = list(STOP_WORDS)
```

```python
!python -m spacy download en_core_web_sm
nlp = spacy.load('en_core_web_sm')
```

```python
doc = nlp(text)
for ent in doc.ents:
    print(f"{ent.text}: {ent.label_}")
```

```python
for token in doc:
    print(f"{token.text}: {token.pos_}")
```

```python
tokens = [token.text for token in doc]
print(tokens)
```

```python
punctuation = punctuation + '\n'
punctuation
```

```python
word_frequencies = {}
for word in doc:
  if word.text.lower() not in stopwords:
    if word.text.lower() not in punctuation:
      if word.text not in word_frequencies.keys():
        word_frequencies[word.text] = 1
      else:
        word_frequencies[word.text] += 1
```

```python
print(word_frequencies)
```

```python
max_frequency = max(word_frequencies.values())
```

```python
max_frequency
```

```python
for word in word_frequencies.keys():
  word_frequencies[word] = word_frequencies[word]/max_frequency
```

```python
print(word_frequencies)
```

```python
sentence_tokens = [sent for sent in doc.sents]
print(sentence_tokens)
```

```python
sentence_scores = {}
for sent in sentence_tokens:
  for word in sent:
    if word.text.lower() in word_frequencies.keys():
      if sent not in sentence_scores.keys():
        sentence_scores[sent] = word_frequencies[word.text.lower()]
      else:
        sentence_scores[sent] += word_frequencies[word.text.lower()]
```

```python
sentence_scores
```

```python
from heapq import nlargest
```

```python
print(len(sentence_tokens))
select_length = int(len(sentence_tokens)*0.3)
select_length
```

```python
summary = nlargest(select_length, sentence_scores, key = sentence_scores.get)
```

```python
summary
```

```python
final_summary = [word.text for word in summary]

```

```python
summary = ' '.join(final_summary)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer

# Tokenize the summaries
tokenized_summaries = [final_summary, summary]

# Initialize CountVectorizer
vectorizer = CountVectorizer().fit_transform(tokenized_summaries)

# Calculate cosine similarity
cosine_similarities = cosine_similarity(vectorizer)

# Print cosine similarity
print("Cosine Similarity between Final Summary and Reference Summary:", cosine_similarities[0, 1])
```
